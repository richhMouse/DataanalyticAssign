1. Why dictionary keys must be immutable in Python

Reason:-
Dictionary keys are stored using a hash value.
If a key changes after insertion, its hash changes.
Then Python cannot find the key again. That breaks the dictionary.



```python
# Example
# # Valid key
d = {}
d[10] = "value"
d["name"] = "Raja"
print(d)
```

```python
# Invalid key
d = {}
d[[1, 2, 3]] = "value"   # list is mutable
```

2. Program to count frequency of elements in a list

Logic:-

Traverse the list
Use dictionary
Increment count if element exists
Else initialize with 1

```python
elements = [1, 2, 2, 3, 1, 4, 2, 3]

freq = {}

for item in elements:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)
```

3. Difference between dict.get(key) and dict[key]

Difference:-

Method	                         Behavior if key missing

dict[key]	                        Raises KeyError
dict.get(key)	               Returns None or default value

```python
student = {"name": "Raja", "age": 20}

print(student["marks"])
```

```python
student = {"name": "Raja", "age": 20}

marks = student.get("marks", 0)
print(marks)
```

4. Merge two dictionaries and keep larger value for common keys

Logic:-

Copy first dictionary
Traverse second dictionary
If key exists, compare values
Keep the larger one

```python
d1 = {"a": 10, "b": 20, "c": 30}
d2 = {"b": 25, "c": 15, "d": 40}

result = d1.copy()

for key, value in d2.items():
    if key in result:
        result[key] = max(result[key], value)
    else:
        result[key] = value

print(result)
```

5. Dictionary comprehension for cubes of odd numbers only

Condition:-

Input is a list of integers
Keep only odd numbers
Value is cube of number

```python
numbers = [1, 2, 3, 4, 5, 6, 7]

cube_dict = {n: n**3 for n in numbers if n % 2 != 0}

print(cube_dict)
```
